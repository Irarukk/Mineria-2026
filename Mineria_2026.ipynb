##KPI 1 - Nivel de actualización del catálogo

###Descripción y justificación

Este KPI mide el nivel de actualización del catálogo de contenidos de Netflix, considerando el año de lanzamiento de los títulos disponibles. La vigencia del contenido es un factor clave en la percepción de valor del servicio, ya que los usuarios tienden a preferir contenidos recientes y alineados con las tendencias actuales del mercado audiovisual.

Un catálogo actualizado contribuye a la competitividad de la plataforma, permitiendo responder de mejor manera a la demanda por contenido relevante y apoyando procesos de planificación, adquisición y producción de nuevos títulos.

| Elemento                   | Descripción                                                                                                          |
| -------------------------- | -------------------------------------------------------------------------------------------------------------------- |
| **Nombre del KPI**         | Nivel de actualización del catálogo                                                                                  |
| **Objetivo asociado**      | Evaluar la vigencia del contenido disponible en Netflix                                                              |
| **Descripción**            | Mide el porcentaje de títulos del catálogo cuyo año de lanzamiento se encuentra dentro de un rango reciente definido |
| **Indicador**              | Porcentaje de contenidos recientes                                                                                   |
| **Fórmula**                | (Cantidad de títulos con `release_year` ≥ Año actual − X / Total de títulos del catálogo) × 100                      |
| **Unidad de medida**       | Porcentaje (%)                                                                                                       |
| **Criterio de medición**   | Contenidos lanzados dentro de los últimos *X* años                                                                   |
| **Frecuencia de medición** | Anual                                                                                                                |
| **Fuente de datos**        | Dataset Netflix (`release_year`)                                                                                     |
| **Responsable**            | Área de Análisis de Datos                                                                                            |
| **Meta**                   | ≥ 60% de contenidos recientes                                                                                        |
| **Interpretación**         | Valores altos indican un catálogo actualizado y competitivo                                                          |

## KPI 2 – Popularidad promedio del catálogo de Netflix

###Descripción y justificación

Este KPI mide la proporción de contenidos del catálogo que se encuentran dentro del percentil superior de popularidad del mercado (TMDB). Permite evaluar si Netflix concentra contenido con alto atractivo comercial, evitando el uso del promedio y reduciendo el impacto de valores extremos.

Este indicador se alinea directamente con el OG, ya que permite comprender el posicionamiento del catálogo en relación con el mercado.

| Elemento                   | Descripción                                                                         |
| -------------------------- | ----------------------------------------------------------------------------------- |
| **Nombre KPI**             | Porcentaje de contenido altamente popular                                           |
| **Objetivo**               | Evaluar el nivel de atractivo del catálogo                                          |
| **Descripción**            | Proporción de títulos del catálogo ubicados en el percentil superior de popularidad |
| **Fórmula**                | (Títulos con popularity ≥ P75 del mercado / Total de títulos del catálogo) × 100    |
| **Unidad de medida**       | Porcentaje (%)                                                                      |
| **Frecuencia de medición** | Mensual                                                                             |
| **Fuente de datos**        | Dataset TMDB                                                                        |
| **Responsable**            | Área de Analítica / Contenidos                                                      |
| **Meta**                   | ≥ 25%                                                                               |
| **Interpretación**         | Valores altos indican mayor concentración de contenido atractivo                    |

## KPI 3 – Valoración del catálogo de Netflix

###Descripción y justificación

Este KPI mide la proporción de contenidos del catálogo que superan un umbral mínimo de valoración, permitiendo evaluar la calidad percibida del contenido sin recurrir a promedios. Este enfoque es más robusto y alineado con prácticas de BI.

El indicador permite comprender si el catálogo está compuesto mayoritariamente por títulos bien evaluados por los usuarios, apoyando decisiones de planificación y gestión.

| Elemento                   | Descripción                                                            |
| -------------------------- | ---------------------------------------------------------------------- |
| **Nombre KPI**             | Proporción de contenido bien valorado                                  |
| **Objetivo**               | Evaluar la calidad percibida del catálogo                              |
| **Descripción**            | Porcentaje de títulos con valoración superior al umbral definido       |
| **Fórmula**                | (Títulos con vote_average ≥ 6.5 / Total de títulos del catálogo) × 100 |
| **Unidad de medida**       | Porcentaje (%)                                                         |
| **Frecuencia de medición** | Mensual                                                                |
| **Fuente de datos**        | Dataset TMDB                                                           |
| **Responsable**            | Área de Contenidos                                                     |
| **Meta**                   | ≥ 60%                                                                  |
| **Interpretación**         | Valores altos reflejan buena aceptación del catálogo                   |

```python
from datetime import datetime

X = 5  # puedes cambiar a 3, 5 o 7 según tu criterio
anio_actual = datetime.now().year

kpi1_total = df_netflix["release_year"].notna().sum()
kpi1_recientes = (df_netflix["release_year"] >= (anio_actual - X)).sum()
kpi1_pct = (kpi1_recientes / kpi1_total) * 100 if kpi1_total > 0 else 0

print(f"KPI 1 - Nivel de actualización (últimos {X} años): {kpi1_pct:.2f}%")
```

```python
# Percentil 75 del mercado (TMDB)
p75_pop = df_tmbd["popularity"].dropna().quantile(0.75)

# Catálogo Netflix enriquecido (solo filas con popularity disponible)
cat = df_final[df_final["popularity"].notna()].copy()

kpi2_total = len(cat)
kpi2_high = (cat["popularity"] >= p75_pop).sum()
kpi2_pct = (kpi2_high / kpi2_total) * 100 if kpi2_total > 0 else 0

print(f"KPI 2 - % contenido altamente popular (>= P75 TMDB): {kpi2_pct:.2f}%")
print("P75 popularity (TMDB):", p75_pop)
```

```python
cat_votes = df_final[df_final["vote_average"].notna() & df_final["vote_count"].notna()].copy()

kpi3_total = len(cat_votes)
kpi3_good = ((cat_votes["vote_average"] >= 6.5) & (cat_votes["vote_count"] >= 500)).sum()
kpi3_pct = (kpi3_good / kpi3_total) * 100 if kpi3_total > 0 else 0

print(f"KPI 3 - % contenido bien valorado (>=6.5 y >=500 votos): {kpi3_pct:.2f}%")
```

```python
p75_pop_market = df_tmbd["popularity"].dropna().quantile(0.75)

market = df_tmbd.copy()
cond = (
    (market["popularity"].notna()) &
    (market["vote_average"].notna()) &
    (market["vote_count"].notna()) &
    (market["popularity"] >= p75_pop_market) &
    (market["vote_average"] >= 6.5) &
    (market["vote_count"] >= 500)
)

kpi4_count = cond.sum()

print("KPI 4 - Volumen de oportunidades de expansión (mercado atractivo):", int(kpi4_count))
print("P75 popularity (TMDB):", p75_pop_market)
```

```python
kpis_resumen = pd.DataFrame({
    "KPI": ["KPI 1", "KPI 2", "KPI 3", "KPI 4"],
    "Nombre": [
        f"Nivel de actualización (últimos {X} años)",
        "% contenido altamente popular (>= P75 TMDB)",
        "% contenido bien valorado (>=6.5 y >=500 votos)",
        "Volumen de oportunidades de expansión (mercado atractivo)"
    ],
    "Resultado": [kpi1_pct, kpi2_pct, kpi3_pct, kpi4_count],
    "Unidad": ["%", "%", "%", "títulos"]
})
kpis_resumen
```
