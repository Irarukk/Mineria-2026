Dataset Final

El dataset final contiene 5.269 películas de Netflix, enriquecidas con métricas de TMDB,
alcanzando un 78,59 % de coincidencia mediante una estrategia de unión en dos etapas.


```python
# Copias
net = df_netflix.copy()
tmdb = df_tmbd.copy()

# 1) Netflix: SOLO películas
net["type"] = net["type"].astype(str)
net_movies = net[net["type"].str.strip().str.lower().isin(["movie", "película", "pelicula"])].copy()

# 2) Normalizador
def normalize_title(s):
    if pd.isna(s):
        return np.nan
    s = str(s).lower().strip()
    s = unicodedata.normalize("NFKD", s).encode("ascii", "ignore").decode("ascii")
    s = re.sub(r"[\(\)\[\]\{\}]", " ", s)
    s = re.sub(r"[^a-z0-9\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

# 3) Keys Netflix
net_movies["title_key"] = net_movies["title"].apply(normalize_title)
net_movies["year_key"]  = pd.to_numeric(net_movies.get("release_year", np.nan), errors="coerce").astype("Int64")

# 4) Keys TMDB (título + año)
# Si TMDB tiene original_title, lo usamos como respaldo después
tmdb["title_key"] = tmdb["title"].apply(normalize_title)

if "release_year" in tmdb.columns:
    tmdb_year_series = tmdb["release_year"]
elif "release_date" in tmdb.columns:
    tmdb_year_series = pd.to_datetime(tmdb["release_date"], errors="coerce").dt.year
elif "year" in tmdb.columns:
    tmdb_year_series = tmdb["year"]
else:
    tmdb_year_series = pd.Series([pd.NA] * len(tmdb), index=tmdb.index)

tmdb["year_key"] = pd.to_numeric(tmdb_year_series, errors="coerce").astype("Int64")

# 5) Preparar TMDB "mejor opción" (por popularity si existe)
sort_col = "popularity" if "popularity" in tmdb.columns else None
tmdb_sorted = tmdb.sort_values(sort_col, ascending=False) if sort_col else tmdb

# A) Dedup estricto (title_key, year_key)
tmdb_best_ty = tmdb_sorted.drop_duplicates(subset=["title_key", "year_key"], keep="first").copy()

# B) Dedup por solo título (title_key) -> para rescatar matches sin año
tmdb_best_t = tmdb_sorted.drop_duplicates(subset=["title_key"], keep="first").copy()

# 6) Merge etapa 1: título + año
m1 = pd.merge(
    net_movies,
    tmdb_best_ty,
    how="left",
    on=["title_key", "year_key"],
    suffixes=("_netflix", "_tmdb")
)

# Identificar columnas TMDB (las que vienen desde tmdb, excluyendo keys)
tmdb_cols = [c for c in tmdb.columns if c not in ["title_key", "year_key"]]

# 7) Merge etapa 2 (rescate): solo título, SOLO para los que no matchearon en etapa 1
# Detectamos "no match" mirando una col típica de TMDB (vote_average o popularity)
probe = "vote_average" if "vote_average" in m1.columns else ("popularity" if "popularity" in m1.columns else None)

if probe is None:
    # Si no hay probe, igual hacemos rescate usando una marca temporal:
    m1["_matched_stage1"] = False
    no_match_mask = pd.Series([True] * len(m1), index=m1.index)
else:
    no_match_mask = m1[probe].isna()

m1_nomatch = m1.loc[no_match_mask, [c for c in m1.columns if c not in tmdb_cols]].copy()

m2 = pd.merge(
    m1_nomatch,
    tmdb_best_t[["title_key"] + tmdb_cols],
    how="left",
    on=["title_key"],
    suffixes=("", "_tmdb2")
)

# 8) Rellenar en m1 con lo encontrado en m2 (solo filas sin match)
m_final = m1.copy()
for c in tmdb_cols:
    if c in m_final.columns and c in m2.columns:
        m_final.loc[no_match_mask, c] = m_final.loc[no_match_mask, c].combine_first(m2[c])

# 9) Métricas
def match_rate(df, col):
    return df[col].notna().mean() * 100

probe2 = "vote_average" if "vote_average" in m_final.columns else ("popularity" if "popularity" in m_final.columns else None)

print("Netflix original:", df_netflix.shape)
print("Netflix solo películas:", net_movies.shape)
print("TMDB original:", df_tmbd.shape)
print("TMDB best (titulo+año):", tmdb_best_ty.shape)
print("TMDB best (solo titulo):", tmdb_best_t.shape)
print("DF final:", m_final.shape)

if probe2:
    stage1 = match_rate(m1, probe2)
    final  = match_rate(m_final, probe2)
    print(f"Match rate etapa 1 (título+año) usando '{probe2}': {stage1:.2f}%")
    print(f"Match rate final (con rescate por solo título) usando '{probe2}': {final:.2f}%")
else:
    print("No encontré vote_average/popularity para medir match rate.")

# Resultado final
df_final = m_final
```

## Tratamiento de valores nulos

Luego de identificar los valores nulos presentes en el dataset, se aplicaron distintas estrategias de limpieza dependiendo del tipo de variable y del porcentaje de datos faltantes.

Para las variables categóricas y de texto se reemplazaron los valores nulos por `"Desconocido"` con el objetivo de conservar la mayor cantidad posible de registros y evitar pérdida innecesaria de información.

En el caso de las variables numéricas relevantes para el análisis, los valores faltantes fueron reemplazados utilizando la mediana, ya que este método reduce el impacto de valores extremos y permite conservar la distribución general de los datos.

```python
# Reemplazo de nulos en variables categóricas y de texto
# Se utiliza "Desconocido" para conservar registros
# y mantener consistencia en el dataset.

categorical_text_columns = [
    'director',
    'cast',
    'country',
    'description',
    'listed_in'
]

for col in categorical_text_columns:
    df_final[col] = df_final[col].fillna('Desconocido')

# Reemplazo de nulos en variables numéricas
# Se utiliza la mediana para evitar el impacto de valores extremos.

numeric_columns = [
    'vote_average',
    'vote_count',
    'release_year',
    'id'
]

for col in numeric_columns:
    df_final[col] = df_final[col].fillna(df_final[col].median())

df_final.isnull().sum()
```

## Resultados del tratamiento de valores nulos

Luego de aplicar las técnicas de limpieza e imputación, se logró reducir significativamente la cantidad de valores faltantes presentes en el dataset.

Para las variables categóricas y de texto se utilizó el valor `"Desconocido"` con el objetivo de conservar la mayor cantidad posible de registros y evitar la pérdida innecesaria de información relevante para el análisis.

En el caso de las variables numéricas, los valores faltantes fueron reemplazados utilizando la mediana, debido a que este método permite reducir el impacto de valores extremos y conservar de mejor manera la distribución general de los datos.

Además, se decidió mantener algunas columnas provenientes del proceso de integración entre Netflix y TMDB, aun cuando presentaban una alta cantidad de valores nulos, ya que estos datos faltantes reflejan la ausencia de coincidencias durante el proceso de merge y no necesariamente errores del dataset original.

# FASE 5 – MODELADO

## Definición del problema

Para esta etapa del proyecto se decidió abordar el problema como una tarea de regresión, debido a que la variable objetivo `vote_average` corresponde a un valor numérico continuo.

El objetivo del modelo será analizar si variables relacionadas con popularidad, interacción del público, duración, presupuesto e ingresos permiten estimar las valoraciones promedio obtenidas por las películas dentro del dataset.

## Selección de variables

En base a los análisis realizados durante el EDA, se seleccionaron variables numéricas que presentan relación con popularidad, desempeño comercial e interacción de los usuarios.

Las variables escogidas para el modelado fueron:

- popularity
- vote_count
- duration
- budget
- revenue
- release_year

Estas variables fueron seleccionadas debido a su relevancia dentro del análisis exploratorio y su posible influencia sobre la valoración promedio de las películas.

## División de los datos

Para entrenar y evaluar los modelos, el dataset fue dividido en conjuntos de entrenamiento y prueba.

Esta separación permite entrenar los modelos utilizando una parte de los datos y posteriormente evaluar su desempeño con información no utilizada durante el entrenamiento.

```python
from sklearn.model_selection import train_test_split

# Variables predictoras
X = df_final[
    [
        "popularity",
        "vote_count",
        "duration",
        "budget",
        "revenue",
        "release_year"
    ]
]

# Variable objetivo
y = df_final["vote_average"]

# Eliminar filas con valores faltantes
data_model = pd.concat([X, y], axis=1).dropna()

X = data_model.drop("vote_average", axis=1)
y = data_model["vote_average"]

# División entrenamiento/prueba
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

print("Datos entrenamiento:", X_train.shape)
print("Datos prueba:", X_test.shape)
```

Luego de eliminar los registros con valores faltantes en las variables seleccionadas para el modelado, el dataset final utilizado quedó compuesto por 4.139 registros.

Posteriormente, los datos fueron divididos en un conjunto de entrenamiento (70%) y un conjunto de prueba (30%). Esto permite que el modelo aprenda patrones utilizando una parte de los datos y posteriormente sea evaluado con registros distintos a los utilizados durante el entrenamiento.

De esta manera, es posible analizar qué tan bien logra generalizar el modelo frente a información que no conocía previamente.

## Modelo 1 — Regresión Lineal

Como primer enfoque de modelado, se utilizó una regresión lineal para analizar la relación entre las variables seleccionadas y la valoración promedio (`vote_average`) de las películas.

Este modelo fue seleccionado debido a que la variable objetivo corresponde a un valor numérico continuo, lo que hace que la regresión lineal sea adecuada para este tipo de problema.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import (
    r2_score,
    mean_absolute_error,
    mean_squared_error,
    root_mean_squared_error
)

import matplotlib.pyplot as plt

# CREACIÓN DEL MODELO

modelo_lr = LinearRegression()

# ENTRENAMIENTO

modelo_lr.fit(X_train, y_train)

# PREDICCIONES

y_pred_lr = modelo_lr.predict(X_test)

# EVALUACIÓN DEL MODELO

r2_lr = r2_score(y_test, y_pred_lr)
mae_lr = mean_absolute_error(y_test, y_pred_lr)
mse_lr = mean_squared_error(y_test, y_pred_lr)
rmse_lr = root_mean_squared_error(y_test, y_pred_lr)

print("R² :", r2_lr)
print("MAE:", mae_lr)
print("MSE:", mse_lr)
print("RMSE:", rmse_lr)

# VISUALIZACIÓN DEL MODELO VALORES REALES VS PREDICHOS

plt.figure(figsize=(7,5))

plt.scatter(
    y_test,
    y_pred_lr,
    alpha=0.6
)

plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    linestyle="--",
    color="red"
)

plt.title("Regresión Lineal: Valores reales vs predichos")
plt.xlabel("Valores reales de vote_average")
plt.ylabel("Valores predichos de vote_average")

plt.grid(True)

plt.show()
```

### Evaluación del modelo

La Regresión Lineal fue evaluada mediante las métricas **R²**, **MAE**, **MSE** y **RMSE**, con el objetivo de medir tanto su capacidad explicativa como el error de sus predicciones.

El valor de **R² = 0.0547** indica que el modelo explica solo una pequeña parte de la variación de `vote_average`, lo que evidencia un bajo poder predictivo. Por su parte, las métricas **MAE = 1.10**, **MSE = 2.87** y **RMSE = 1.69** muestran que las predicciones presentan un nivel de error considerable.

La visualización de los valores reales frente a los valores predichos confirma estos resultados, ya que la mayoría de los puntos se encuentran alejados de la línea de referencia y el modelo tiende a concentrar las predicciones alrededor de un mismo valor.

En conjunto, los resultados indican que la Regresión Lineal no logra representar adecuadamente el comportamiento de `vote_average`, por lo que fue necesario evaluar modelos capaces de capturar relaciones más complejas entre las variables.

## Modelo 2 — Árbol de Decisión

Como segundo enfoque de modelado, se utilizó un Árbol de Decisión para analizar si un modelo no lineal lograba representar mejor el comportamiento de la variable `vote_average`.

A diferencia de la regresión lineal, este modelo puede identificar relaciones más complejas entre las variables, permitiendo dividir los datos en distintos grupos según sus características.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import (
    r2_score,
    mean_absolute_error,
    mean_squared_error,
    root_mean_squared_error
)

import matplotlib.pyplot as plt

# CREACIÓN DEL MODELO

modelo_tree = DecisionTreeRegressor(
    random_state=42,
    max_depth=5
)

# ENTRENAMIENTO

modelo_tree.fit(X_train, y_train)

# PREDICCIONES

y_pred_tree = modelo_tree.predict(X_test)

# EVALUACIÓN DEL MODELO

r2_tree = r2_score(y_test, y_pred_tree)
mae_tree = mean_absolute_error(y_test, y_pred_tree)
mse_tree = mean_squared_error(y_test, y_pred_tree)
rmse_tree = root_mean_squared_error(y_test, y_pred_tree)

print("R² :", r2_tree)
print("MAE:", mae_tree)
print("MSE:", mse_tree)
print("RMSE:", rmse_tree)
```

El Árbol de Decisión fue evaluado mediante las métricas **R²**, **MAE**, **MSE** y **RMSE**, con el objetivo de medir su capacidad predictiva y el error de sus estimaciones.

Además, las métricas **MAE = 0.7667**, **MSE = 1.1210** y **RMSE = 1.0588** evidencian una reducción en el error de predicción.

La visualización de los valores reales frente a los valores predichos confirma este comportamiento, ya que los puntos presentan una mayor cercanía a la línea de referencia en comparación con la Regresión Lineal.

En conjunto, estos resultados indican que el Árbol de Decisión representa de mejor manera los patrones presentes en el conjunto de datos, siendo el modelo con mejor desempeño entre los modelos supervisados evaluados.

## Modelo 3 — K-MEANS

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Transformación logarítmica de popularity
# Se aplica para reducir el efecto de valores extremos.
df_final = df_final.copy()
df_final.loc[:, "popularity_log"] = np.log1p(df_final["popularity"])

# Selección de variables para clustering
cluster_vars = ["popularity_log", "vote_average", "vote_count"]

df_cluster = df_final[cluster_vars].dropna()

# Escalado de variables
scaler = StandardScaler()
X_cluster_scaled = scaler.fit_transform(df_cluster)

df_cluster.head()
```

# MÉTODO DEL CODO

```python
import matplotlib.pyplot as plt

inertia = []

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_cluster_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), inertia, marker="o")

plt.title("Método del Codo")
plt.xlabel("Número de Clusters")
plt.ylabel("Inercia")

plt.grid(True)

plt.show()
```

El Método del Codo fue utilizado para determinar la cantidad óptima de clusters para el algoritmo K-Means.

Como se observa en el gráfico, la disminución de la inercia es considerable hasta llegar a **3 clusters**. A partir de ese punto, la reducción comienza a ser menos significativa, por lo que se seleccionó **k = 3** como el número más adecuado para realizar la segmentación de las películas.

```python
# ==========================================
# ENTRENAMIENTO DEL MODELO K-MEANS
# ==========================================

kmeans = KMeans(
    n_clusters=3,
    random_state=42,
    n_init=10
)

df_cluster["Cluster"] = kmeans.fit_predict(X_cluster_scaled)

df_cluster.head()
```

El gráfico muestra la segmentación realizada por K-Means utilizando las variables `popularity_log`, `vote_average` y `vote_count`.

Se identifican tres grupos principales. El primer grupo concentra la mayor parte de las películas con niveles de popularidad bajos o medios y valoraciones intermedias. El segundo grupo reúne películas con mayores niveles de popularidad respecto al resto del conjunto de datos y valoraciones promedio superiores. El tercer grupo está compuesto principalmente por títulos con baja popularidad y bajas valoraciones.
